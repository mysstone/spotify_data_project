```python
stream_final_df_gb = stream_final_df_gb[(stream_final_df_gb['year'] > 2013) 
| ((stream_final_df_gb['year'] == 2013) & (stream_final_df_gb['month'] >= 9))]

stream_final_df_gb
```

```python
plt.figure(figsize=(12,6))
plt.plot(stream_final_df_gb['date'], stream_final_df_gb['hours_played'], label="Total hours_played per month", alpha=0.5)
plt.plot(stream_final_df_gb['date'], stream_final_df_gb['rolling_avg'], label="3-Month Rolling Average", linewidth=2, color='red')

plt.xlabel("Date")
plt.ylabel("Total Listening Time (hours)")
plt.title("Monthly Listening Trends")
plt.legend()
plt.show()
```
